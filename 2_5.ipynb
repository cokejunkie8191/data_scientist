ライブラリのインポート

```
# MatplotliとSeabornの読み込み
# Seabornはきれいに図示できる
import numpy as np
import matplotlib as mpl
import seaborn as sns

# pyplotにはpltの別名で実行できるようにする
import matplotlib.pyplot as plt

# Jupyter Notebook上でグラフを表示させるために必要なマジックコマンド
%matplotlib inline
```

```
散布図
```

```
# シード値の固定
np.random.seed(0)

# x軸のデータ
x = np.random.randn(30)

# y軸のデータ
y = np.random.randn(30)

# グラフの大きさ指定
plt.figure(figsize=(20, 6))

# グラフの描画
plt.plot(x, y, 'o')

# タイトル
plt.title('Title Name')
# xの座標名
plt.xlabel('X')
# yの座標名
plt.ylabel('Y')

# gridの表示
plt.grid(True)
```

```
# 連続曲線

# シード値の固定
np.random.seed(0)

# x軸のデータ
numpy_data_x = np.arange(1000)

# y軸のデータ
numpy_random_data_y = np.random.randn(1000).cumsum()

# グラフの大きさ指定
plt.figure(figsize=(20, 6))

# グラフの描画
plt.plot(numpy_data_x, numpy_random_data_y, label='Label')
plt.legend()

# タイトル
plt.title('Title Name')
# xの座標名
plt.xlabel('X')
# yの座標名
plt.ylabel('Y')

# gridの表示
plt.grid(True)
```

グラフの分割

```
# グラフの大きさを固定
plt.figure(figsize=(20, 6))

# 2行1列のグラフの1つ目
plt.subplot(2, 1, 1)
x = np.linspace(-10, 10, 100)
plt.plot(x, np.sin(x))

# 2行2列のグラフの2つ目
plt.subplot(2, 1, 2)
y = np.linspace(-10, 10, 100)
plt.plot(y, np.sin(2 * y))

plt.grid(True)
```

関数グラフの描画

```
def my_function(x):
  return x ** 2 + 2 * x + 1

x = np.arange(-10, 10)
plt.figure(figsize = (20, 6))
plt.plot(x, my_function(x))
plt.grid(True)
```

ヒストグラム

```
# シードの固定
np.random.seed(0)

# グラフの大きさ指定
plt.figure(figsize = (20, 6))

# ヒストグラムの描画
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins = 60, range = (20, 80))

plt.grid(True)
```
